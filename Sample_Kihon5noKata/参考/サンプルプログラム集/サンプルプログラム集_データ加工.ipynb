## **データのダウンロード**

```python
import seaborn as sns
import pandas as pd
df = sns.load_dataset('titanic')

display(df.head())
```

## **データ情報の確認**

```python
df.info()
```

## **縦に結合（ユニオン）**

```python
df_1 = df.copy()
df_2 = df.copy()
print(len(df_1))
print(len(df_2))
```

```python
df_all = pd.concat([df_1, df_2],ignore_index=True)
print(len(df_all)) #df_1とdf_2を縦に結合したので２倍のデータ件数となります
```

## **欠損値処理（補間）**

```python
df.isna().sum()
```

ageを0で補間

```python
df_na = df.copy()
df_na['age'].fillna(0, inplace=True)
df_na.isna().sum()
```

ageを平均値で補間

```python
df_na = df.copy()
age_avg = df['age'].mean()
df_na['age'].fillna(age_avg, inplace=True)
df_na.isna().sum()
```

## **型の変換**

```python
df.info()
```

```python
# int→str
df['pclass'] = df['pclass'].astype(str)
df.info()
```

```python
# str→float
df['pclass'] = df['pclass'].astype(float)
df.info()
```

```python
# float→int
df['pclass'] = df['pclass'].astype(int)
df.info()
```

## **文字列の置換**

```python
# femaleをwomanに置換
df['sex'].str.replace('female','woman')
```

```python
# 小文字→大文字へ変換
df['sex'].str.upper()
```

```python
# 大文字→小文字へ変換
df['class'].str.lower()
```

## **データの抽出**

```python
# 「who」列の値に'w'を含む行の抽出
df.loc[df['who'].str.contains('w')]
```

```python
# 「embarked」列の値が'C'または'Q'の行の抽出
df.loc[(df['embarked']=='C')|(df['embarked']=='Q')]
```

```python
# 「embarked」列の値が'C'または'Q'の行の抽出（別パターン）
tg_embarked = ['C','Q']
df.loc[df['embarked'].isin(tg_embarked)]
```

```python
# 「age」列の値が10以上の行の「survived」「pclass」「age」列を抽出
df.loc[df['age']>=10,['survived','pclass','age']]
```

## **列の削除**

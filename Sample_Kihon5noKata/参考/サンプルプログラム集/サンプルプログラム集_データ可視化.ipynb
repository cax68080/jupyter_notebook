主に可視化で使用するサンプルプログラム集となります。   
サンプルデータ(タイタニックデータ)をダウンロードして進めます。  
ここでは可視化ライブラリseabornを使います。

## **データのダウンロード**

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

display(df.head())
```

## **ヒストグラム**

```python
sns.distplot(df['fare'])
```

```python
sns.histplot(df['fare'])
```

## **棒グラフ**

```python
df_viz = df[['pclass']].copy()
df_viz.loc[:,'count'] = 1
df_viz = df_viz.groupby('pclass',as_index=False).sum()
sns.barplot(x=[1,2,3], y=df_viz['count'])
```

```python
sns.catplot(x="pclass", data=df, kind="count")
```

## **散布図**

```python
df_viz = df[['age','fare','sibsp','parch']]
sns.pairplot(df_viz)
```

```python
sns.scatterplot(x='age', y='fare', data=df)
```

```python
sns.jointplot(x='age', y='fare', data=df)
```

## **箱ひげ図**

```python
sns.catplot(x='pclass', y='age', data=df, kind='box', hue='survived')
```

```python
sns.violinplot(x='pclass', y='age', data=df) 
```

## **ヒートマップ**

```python
df_viz = df.pivot_table(index='pclass', columns='sex', values='survived',aggfunc=sum)
sns.heatmap(df_viz, annot=True)
```

## **折れ線グラフ**

```python
df_viz = df[['age','survived']].copy()
df_viz.loc[:,'count'] = 1
df_viz = df_viz.groupby(['age','survived'],as_index=False).sum()
sns.relplot(x='age', y='count', data=df_viz, kind='line', hue='survived')
```

```python
sns.lineplot(x='age', y='count', data=df_viz, hue='survived')
```

```python
import pandas as pd
```

```python
data_dir = "..\data"
input_dir = f"{data_dir}\input"
output_dir = f"{data_dir}\output"
transaction_file = f"{input_dir}\\transaction.csv"
customer_file = f"{input_dir}\customer_master.csv"
product_file = f"{input_dir}\product_master.csv"
output_file = f"{output_dir}\output.csv"
```

```python
df_tran = pd.read_csv(transaction_file)
df_product = pd.read_csv(product_file)
df_customer = pd.read_csv(customer_file)
display(df_tran.head())
display(df_product.head())
display(df_customer.head())
```

```python
print(df_tran.info())
print(df_tran.shape)
print(df_product.info())
print(df_customer.info())
```

```python
# 欠損値の確認
print(df_tran.isnull().sum())
print(df_product.isnull().sum())
print(df_customer.isnull().sum())
```

```python
display(df_tran.loc[df_tran["product_id"].isnull()])
```

```python
display(df_tran.loc[df_tran["cancellation_date"].notna()].head())
```

```python
print(df_tran.loc[df_tran["cancellation_date"].notna(),"status"].unique())
```

```python
display(df_product.loc[df_product["color"].isnull()])
```

```python
print(df_product.loc[df_product["brand"]=="ディスプレイ","color"].unique())
print(df_product.loc[df_product["brand"]=="マウス","color"].unique())
print(df_product.loc[df_product["brand"]=="キーボード","color"].unique())
```

```python
# 欠損値の除去
print(df_tran.info())
df_tran.dropna(subset=["product_id"],inplace=True)
print(df_tran.info())
print(df_tran.isnull().sum())
```

```python
# 不要な列を削除

df_product_2 = df_product[["product_id","product_name","brand","category","color","price"]]
print(df_product.info())
print(df_product_2.info())
```

```python
# テーブルの結合
df_merge = pd.merge(df_tran,df_product_2,on="product_id",how="left")
display(df_merge.head())
df_merge = pd.merge(df_merge,df_customer,on="customer_id",how="left")
display(df_merge.head())
```

```python
# データをCSVに出力
df_merge.to_csv(output_file,index=False)
```

## **1. ライブラリのインポート**

```python
import pandas as pd
```

## **2. 初期定義**

```python
data_dir = '../data' # このipynbファイルから見たdataフォルダのパス
input_dir = f'{data_dir}/input' # 変数data_dirから見たinputフォルダのパス
output_dir = f'{data_dir}/output'
transaction_file = f'{input_dir}/transaction.csv'
customer_file = f'{input_dir}/customer_master.csv'
product_file = f'{input_dir}/product_master.csv'
output_file = f'{output_dir}/transaction_processed.csv' # 加工済みデータのパス
```

## **3. データの読み込み**

```python
# csvファイルの読み込み
df_transaction = pd.read_csv(transaction_file)
df_customer = pd.read_csv(customer_file)
df_product = pd.read_csv(product_file)
```

## **4. メイン処理 ～データ加工～**

#### **① データの状態を確認しよう**



##### **読み込んだデータを表示してみよう**

```python
# データを表示してみる
display(df_transaction)
display(df_customer)
display(df_product)
```

##### **欠損値の有無を確認しよう**

```python
print(df_transaction.isnull().sum())
```

```python
print(df_customer.isnull().sum())
print(df_product.isnull().sum())
```

##### **欠損値の内容を確認しよう**

```python
display(df_transaction.loc[df_transaction["product_id"].isnull()])
```

```python
display(df_product.loc[df_product["color"].isnull()])
```

##### **重複の有無を確認しよう**

```python
print(df_transaction.duplicated(subset=['transaction_id']).sum())
```

```python
print(df_customer.duplicated(subset=['customer_id']).sum())
print(df_product.duplicated(subset=['product_id']).sum())
```

##### **重複の内容を確認しよう**

```python
display(df_transaction.loc[df_transaction.duplicated(subset=['transaction_id'],keep=False)].sort_values(['transaction_id']))
```

#### **② データを加工しよう** & **③ 加工結果を確認しよう**

##### **明らかに不要なデータを除外しよう**

```python
df_product.drop(["Unnamed: 6"],axis=1,inplace=True)
```

##### **処理結果を確認しよう**

```python
display(df_product)
```

##### **欠損値を除去しよう**

```python
df_transaction.dropna(subset=['product_id'],inplace=True)
```

##### **処理結果を確認しよう**

```python
print(df_transaction.isnull().sum())
```

##### **重複をなくそう**

```python
df_transaction.drop_duplicates(subset=['transaction_id'], keep='last', inplace=True)
```

##### **処理結果を確認しよう**

```python
print(df_transaction.duplicated(subset=['transaction_id']).sum())
```

##### **データを結合しよう**

```python
df_merge = pd.merge(df_transaction,df_customer,on='customer_id',how='left')
```

```python
df_merge = pd.merge(df_merge,df_product,on='product_id',how='left')
```

##### **処理結果を確認しよう**

```python
display(df_merge)
```

```python
print(df_merge.duplicated(subset='transaction_id').sum())
```

```python
print(df_merge.isnull().sum())
```

## **5. データの出力**

```python
df_merge.to_csv(output_file,index=False)
```

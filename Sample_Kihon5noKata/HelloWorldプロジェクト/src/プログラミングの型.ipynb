```python
# 初期定義
data_dir = "../data"
input_dir = f"{data_dir}/input"
output_dir = f"{data_dir}/output"
input_file = f"{input_dir}/気象データ.csv"
output_file = f"{output_dir}/気象データ_加工済.csv"
```

```python
import pandas as pd
```

```python
df = pd.read_csv(input_file)
display(df.head())
```

```python
df["降水有無"] = "無"
df.loc[df["降水量の合計(mm)"] > 0,"降水有無"] = "有"
display(df.head(10))
```

```python
df.to_csv(output_file,index=False)
```

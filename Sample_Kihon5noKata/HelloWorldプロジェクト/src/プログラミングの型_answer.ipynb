## **1. ライブラリのインポート**

```python
import pandas as pd
```

## **2. 初期定義**

```python
data_dir ='../data' # このipynbファイルから見たdataフォルダのパス
input_dir = f'{data_dir}/input' # 変数data_dirから見たinputフォルダのパス
output_dir = f'{data_dir}/output'
input_file = f'{input_dir}/気象データ.csv'
output_file = f'{output_dir}/気象データ_加工済.csv'
```

## **3. データの読み込み**

```python
# csvファイルの読み込み
df = pd.read_csv(input_file)
```

```python
display(df.head(10))
```

## **4. メイン処理**

```python
# 年月日カラムから年だけを抽出する
df['降水有無'] = '無'
df.loc[df['降水量の合計(mm)']>0,'降水有無'] = '有'

display(df.head(10)) # 処理結果の確認
```

## **5. データ出力**

```python
# 出力
df.to_csv(output_file,index=False)
```

## 2.1 アイリスデータセット

### 2.1.1 可視化からはじめる

```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

# sklearnからload_iris関数を用いてデータをロードする
data = load_iris()

features = data['data']
feature_names = data['feature_names']
target = data['target']
target_names = data['target_names']
labels = target_names[target]

for t,marker,c in zip(range(3),">ox","rgb"):
    # クラスごとに色の異なるマーカーでプロットする
    plt.scatter(features[target == t,0],features[target == t,1],marker = marker,c = c)
plt.show()
```

### 2.1.2 はじめての分類モデル作成

```python
# 「花弁の長さ」は配列の3番目(インデックス操作では２)に格納されている。
plength = features[:,2]
is_setosa = (labels == 'setosa')

# 次が重要なステップ
max_setosa = plength[is_setosa].max()
min_non_setosa = plength[~is_setosa].min()
print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))
```

```python
# 花弁の長さからSetosaとSetosa以外を分類するモデル
def apply_model(example):
    if example[2] < 2:
        print('Iris Setosa')
    else:
        print('Iris Virginica or Iris Versicolor')
        
```

```python
# VirginicaとVersicolorの分類
# Setosa以外の特徴量とラベルを選ぶ
features = features[~is_setosa]
labels = labels[~is_setosa]
virginica = (labels == 'virginica')
```

```python
best_acc = -1.0
best_fi = -1.0
best_t = -1.0

for fi in range(features.shape[1]):
    # 各特徴量ごとに閾値の候補を生成する
    thresh = features[:,fi].copy()
    thresh.sort()
    #print(thresh)
    
    # すべての閾値でテストする
    for t in thresh:
        pred = (features[:,fi] > t)
        #print(pred)
        acc = (labels[pred] == 'virginica').mean()
        print(acc)
        if acc > best_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
```

```python
def fix_unicode(text):
    return text.replace(u"\u2019",",")
```

```python
from bs4 import BeautifulSoup
import requests
import re
url = "http://radar.oreilly.com/2010/06/what-is-data-science.html"
#url = "https://business.nikkei.com/atcl/gen/19/00322/072000053/"
html = requests.get(url).text
soup = BeautifulSoup(html,'html5lib')
content = soup.find("div","entry-content")
regex = r"[\w']+|[\.]"

document = []

for paragraph in content("p"):
    words = re.findall(regex,fix_unicode(paragraph.text))
    document.extend(words)
```

```python
print(document)
```

```python
from collections import defaultdict

bigrams = zip(document,document[1:])
transitions = defaultdict(list)
for prev,current in bigrams:
    transitions[prev].append(current)
```

```python
print(transitions)
```

```python
def generate_using_bigrams():
    import random
    current = "."
    result = []
    
    while True:
        next_word_candidates = transitions[current]
       # print(next_word_candidates)
        current = random.choice(next_word_candidates)
        result.append(current)
        if current == ".":
            return " ".join(result)
```

```python
data_txt = generate_using_bigrams()
print(data_txt)
```

```python
trigrams = zip(document,document[1:],document[2:])
trigram_transitions = defaultdict(list)
starts=[]

for prev,current,next in trigrams:
    if prev == ".":
        starts.append(current)
    trigram_transitions[(prev,current)].append(next)
    
```

```python
print( trigram_transitions)
```

```python
def generate_using_trigrams():
    import random
    current = random.choice(starts)
    prev = "."
    
    result = [current]
    while True:
        next_word_cadidates = trigram_transitions[(prev,current)]
        next_word = random.choice(next_word_cadidates)
        prev,current = current,next_word
        result.append(current)
        
        if current == ".":
            return " ".join(result)
```

```python
tri_text = generate_using_trigrams()
print(tri_text)
```
